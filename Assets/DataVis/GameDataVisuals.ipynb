```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from matplotlib.collections import LineCollection
```

```python
df = pd.read_csv("../SavedData/gameData.csv")
df.head()
```

```python
plt.figure()
plt.plot(df["time"], df["posX"])
plt.xlabel("Time")
plt.ylabel("posX")
plt.title("Player X Position Over Time")
plt.show()
```
